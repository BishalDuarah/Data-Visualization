```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['age'], df['chol'], color='green')
plt.title('Line Graph: Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```

```python
age_groups = df.groupby('age')['chol'].mean().head(10)
age_groups.plot(kind='bar', figsize=(10, 5), color='orange')
plt.title('Bar Graph: Avg Cholesterol by Age')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='age', y='chol', data=df)
plt.title('Scatter Plot: Age vs Cholesterol')
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(df['chol'], fill=True, color='purple')
plt.title('Density Plot: Cholesterol Distribution')
plt.xlabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(y='chol', data=df, color='cyan')
plt.title('Violin Plot: Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['age', 'chol']])
plt.title('Box Plot: Age and Cholesterol')
plt.ylabel('Value')
plt.xlabel('Features')
plt.show()
```
